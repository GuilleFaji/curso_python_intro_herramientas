Lo primero vamos a leer los datos del fichero air_quality_no2.csv de la carpeta data

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
air_quality = pd.read_csv("data/air_quality_no2.csv", index_col=0, parse_dates=True)
air_quality.head()
```

El método plot de pandas nos permite hacer gráficas tanto de Series como de DataFrames

```python
air_quality.plot(figsize = (16,9))
```

Para poder ver mejor los datos, graficamos una de las columnas

Podemos crear distintos tipos de gráficas empleando los distintos tipos que tiene plot. Podemos ver un listado con todos los tipos de gráficos soportados:

```python
[
    method_name
    for method_name in dir(air_quality.plot)
    if not method_name.startswith("_")
]
```

Para ver la dispoersión de una variable frente a otra usamoes el gráfico scatter:

```python
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5);
```

Para ver información de la mediana y quartiles, usamos el gráfico 


<left>
<img src="imgs/box_plot.png"  alt="drawing" width="40%"/>
</left>

```python
air_quality.plot.box()
```

Podemos hacer una gráfica separada para cada una de las columnas

```python
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)
```

Los gráficos se pueden personalizar, salvar a un fichero. Cada uno de los gráficos es un objeto de matplotlib.

__Matplotlib__ es la librería más usada para hacer gráficos en python (aunque hay otras librerías que también son útiles)

```python
fig, axs = plt.subplots(figsize=(12, 4))

air_quality.plot.area(ax=axs)

axs.set_ylabel("NO$_2$ concentration")

fig.savefig("imgs/no2_concentrations.png") 
```
