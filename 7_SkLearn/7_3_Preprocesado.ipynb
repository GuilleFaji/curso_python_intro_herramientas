Muchos algoritmos se benefician directa o indirectamente de que hagamos un pretratamiento de los datos ya sea porque podemos eliminar desbalanceos entre las dimensiones de diferentes variables o porque directamente algunos algoritmos esperan inputs estandarizados en alguna forma.

El proceso de estandarizado es bastante sencillo e incluye:

```python
import sklearn
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

<font size = 5>

## **A - Estandarización:**
Establecer media 0 y varianza 1 para las variables.

Estamos asumiendo una distribución gausiana.

```python
from sklearn import datasets
iris = datasets.load_iris()
iris.data
```

```python
from sklearn import preprocessing
escalador = preprocessing.StandardScaler().fit(iris.data)

iris_escalado = escalador.transform(iris.data)
print(iris_escalado.mean(axis=0))
print(iris_escalado.std(axis=0))
iris_escalado
```

```python
plt.hist(iris.data[:,1], bins=30)
```

```python
plt.hist(iris_escalado[:,1], bins=30)
```

<font size = 5>

## **B - Escalado:**
Alternativa a la Estandarización: Distribuir los valores a lo largo de un rango (habitualmente 0,1 o -1,1)

En ocasiones no queremos alterar la distribución de los datos (como hacía el anterior método).

**MinMaxScaler** transforma linealmente los datos, manteniendo la forma de la distribución original. Es muy utilizado en algoritmos como K-nearest neighbors y redes neuronales.

**MaxAbsScaler** transformar en función del valor máximo absoluto de cada caracteristica, siendo este máximo 1. Mantiene el signo de los datos. También conserva la 'esparsidad', que es la caracteristica de algunos datasets en los que muchos de sus valores son 0 o casi 0.

```python
#MinMaxScaler: Escala los datos entre el rango que digamos
escalador = preprocessing.MinMaxScaler((-100,100)).fit(iris.data)
iris_escalado = escalador.transform(iris.data)
print(iris_escalado)
```

```python
plt.hist(iris_escalado[:,1], bins=30);
plt.hist(iris.data[:,1], bins=30);
```

```python
# MaxAbsScaler: 
escalador = preprocessing.MaxAbsScaler().fit(iris.data)
iris_escalado = escalador.transform(iris.data)
print(iris_escalado)
```

```python
plt.hist(iris_escalado[:,1], bins=30);
plt.hist(iris.data[:,1], bins=30);
```
