# **7.7 Principal Component Analysis:**

Reducción de dimensionalidad

```python
import sklearn
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

<font size = 5> 

Técnica utilizada para reducir la dimensionalidad de los datos al mismo tiempo que intentamos preservar el máximo de información posible. Esto se logra intentando codificar las variables ya existentes en otras nuevas que sean ortogonales entre ellas. A estas nuevas variables las conocemos como los **Componentes Principales**.

En el fondo lo que intentamos ver es en qué direcciones los datos tienen mas variabilidad, que normalmente no van a ser los ejes originales, y podemos escoger estas direcciones para tomarlas como nuevos ejes y así seguir explicando una parte importante de la información.

<center><img src="./imgs/pca.png">

```python
from sklearn.datasets import load_iris

# Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

from sklearn.decomposition import PCA

# PCA
pca = PCA(n_components=2)

# Fit 
X_pca = pca.fit_transform(X)

explained_variance_ratio = pca.explained_variance_ratio_

print("Explained Variance Ratio:", explained_variance_ratio)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Iris dataset
iris = load_iris()
X = iris.data
y = iris.target


# PCA object
pca = PCA(n_components=3)

# Fit transform
X_pca = pca.fit_transform(X)

explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)

# Scatter plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap='plasma');
```
